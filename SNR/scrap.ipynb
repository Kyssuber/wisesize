```python
from astropy.io import fits
from astropy.wcs import WCS
from astropy.table import Table
from astropy.coordinates import SkyCoord
from astropy.wcs.utils import skycoord_to_pixel
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import os
homedir=os.getenv("HOME")
```

```python
vf = Table.read(homedir+'/Desktop/galfit_files/VF_WISESIZE_v2.fits')
```

```python
ra_1 = vf[vf['group_name']=='NGC3786_GROUP']['RA']
dec_1 = vf[vf['group_name']=='NGC3786_GROUP']['DEC']
ra_2 = vf[vf['group_name']=='NGC3786_GROUP']['RA']
dec_2 = vf[vf['group_name']=='NGC3786_GROUP']['DEC']
im,head = fits.getdata(homedir+'/vf_html_w1_v2/all_input_fits/NGC3786_GROUP-custom-image-W3.fits',header=True)
```

```python
wcs_w3 = WCS(head)
c1 = SkyCoord(ra_1, dec_1, unit="deg")
c2 = SkyCoord(ra_2, dec_2, unit="deg")
```

```python
#regardless or c1 or c2 input, will output xs and ys from left to right
xs, ys = skycoord_to_pixel(c1, wcs_w3)
```

```python
plt.figure(figsize=(3.5,3.5))
plt.imshow(np.log(im),origin='lower')
plt.scatter(xs[0],ys[0],s=5,color='r')
plt.scatter(xs[1],ys[1],s=5,color='r')
plt.show()
```
